```python
def process_dataset(file_path):
    data = json.load(open(file_path))
    for item in data:
        sentence = item["sentence"]
        annotated = model.annotate_text(sentence)[0]
        segmented = " ".join([w["wordForm"] for w in annotated])
        item["segmented_sentence"] = segmented
    return data
```

```python
import torch
from transformers import AutoModel, AutoTokenizer

phobert = AutoModel.from_pretrained("vinai/phobert-base-v2")
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base-v2")
```

```python
def get_embedding(sent):
    input_ids = tokenizer.encode(sent, return_tensors='pt')
    with torch.no_grad():
        output = phobert(input_ids)
    embedding = output.last_hidden_state.mean(dim=1).squeeze()
    return embedding.numpy()
```

```python
X_train = []
y_train = []
for item in train_data:
    segmented = item["segmented_sentence"]
    vec = get_embedding(segmented)
    X_train.append(vec)
    y_train.append(item["sentiment"]) 
```

```python
X_test = []
y_test = []
for item in test_data:
    segmented = item["segmented_sentence"]
    vec = get_embedding(segmented)
    X_test.append(vec)
    y_test.append(item["sentiment"]) 

X_test = np.array(X_test)
y_test = np.array(y_test)
```

# Bài 3: (6 điểm)
Sử dụng mô hình Gaussian Mixture xây dựng 3 clusters cho 3 nhãn 
Positive, Neutral, và Negative với phương pháp tối ưu Expectation Maximization (EM). 
Đánh giá kết quả của mô hình trên tập test sử dụng độ đo Precision, Recall, và F1. 

## Khởi tạo mô hình GMM_EM bằng numpy

```python
import numpy as np

class GMM_EM:
    def __init__(self, n_components=3, max_iter=100, tol=1e-4):
        self.k = n_components
        self.max_iter = max_iter
        self.tol = tol

    def _log_gaussian(self, X, mean, cov):
        n_features = X.shape[1]
        cov += 1e-6 * np.eye(n_features)  # regularization to avoid singular matrix
        cov_det = np.linalg.det(cov)
        cov_inv = np.linalg.inv(cov)
        diff = X - mean
        mahal = np.sum(diff @ cov_inv * diff, axis=1)

        log_prob = -0.5 * (mahal + n_features * np.log(2 * np.pi) + np.log(cov_det))
        return log_prob

    def fit(self, X):
        n_samples, n_features = X.shape
        rng = np.random.default_rng()
        
        # Initialize means, covariances, and weights
        self.means = X[rng.choice(n_samples, self.k, replace=False)]
        self.covs = np.array([np.cov(X.T) for _ in range(self.k)])
        self.weights = np.ones(self.k) / self.k

        log_likelihood_prev = None

        for iteration in range(self.max_iter):
            # E-step: compute responsibilities in log-space
            log_prob_matrix = np.array([
                np.log(self.weights[i]) + self._log_gaussian(X, self.means[i], self.covs[i])
                for i in range(self.k)
            ]).T  # shape: (n_samples, k)

            # log-sum-exp trick for numerical stability
            log_sum = np.logaddexp.reduce(log_prob_matrix, axis=1, keepdims=True)
            log_gamma = log_prob_matrix - log_sum
            gamma = np.exp(log_gamma)

            # M-step
            N_k = np.sum(gamma, axis=0)

            self.means = (gamma.T @ X) / N_k[:, np.newaxis]
            self.covs = np.array([
                ((gamma[:, i][:, np.newaxis] * (X - self.means[i])).T @ (X - self.means[i])) / N_k[i]
                + 1e-6 * np.eye(n_features)
                for i in range(self.k)
            ])
            self.weights = N_k / n_samples

            # Compute log-likelihood
            log_likelihood = np.sum(log_sum)
            if log_likelihood_prev is not None and np.abs(log_likelihood - log_likelihood_prev) < self.tol:
                break
            log_likelihood_prev = log_likelihood

    def predict(self, X):
        log_prob_matrix = np.array([
            np.log(self.weights[i]) + self._log_gaussian(X, self.means[i], self.covs[i])
            for i in range(self.k)
        ]).T  # shape: (n_samples, k)

        return np.argmax(log_prob_matrix, axis=1)
```

## Map giá trị của nhãn sang số 

```python
# B1. Map nhãn thành số
label_map = {'positive': 0, 'neutral': 1, 'negative': 2}
y_true = np.array([label_map[label] for label in y_train])
```

```python
y_true_test = np.array([label_map[label] for label in y_test])
```

## Scale và giảm chiều các giá trị của tập train và test để không bị tràn số

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

# Bước 2: PCA để giảm chiều, ví dụ còn 50 chiều
pca = PCA(n_components=50, random_state=42)
X_reduced = pca.fit_transform(X_train_scaled)
```

```python
X_test_reduced = pca.transform(X_test_scaled)
```

```python
# Bước 3: Huấn luyện GMM
model = GMM_EM(n_components=3)
model.fit(X_reduced)

# Dự đoán
pred_labels = model.predict(X_reduced)
```

```python
from scipy.stats import mode
from sklearn.metrics import accuracy_score
```

```python
# B4. Ánh xạ cụm -> cảm xúc dựa trên majority vote
final_labels = np.zeros_like(pred_labels)
for i in range(model.k):
    mask = pred_labels == i
    if np.sum(mask) == 0:
        continue
    final_labels[mask] = mode(y_true[mask])[0]

# B5. Đánh giá
acc = accuracy_score(y_true, final_labels)
print("Accuracy after cluster-label mapping:", acc)

# print("Log-likelihood history:", model.log_likelihoods)
```
